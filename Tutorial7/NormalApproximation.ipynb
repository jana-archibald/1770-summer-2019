## What is the probability of getting fewer than 10 wins in a sample of size 100.

### Option 1 compute the exact value!

```r
pbinom(9,100,0.167)
```

```r
barplot(dbinom(c(0:30),100,0.167),names.arg=c(0:30))
```

### Option 2 use the normal approximation.

First we need to compute $\mu_{\hat{p}}$ and $\sigma_{\hat{p}}$

$\mu_{\hat{p}}=p=0.167$

$\sigma_{\hat{p}}=\sqrt{\frac{p(1-p)}{n}}=\sqrt{\frac{(0.167)(0.833)}{100}}=0.0372975870533202$

```r
sqrt((0.167*0.833/100))
```

Now we can use pnorm to find the area in the tail.

```r
pnorm(0.10,0.167,0.0372975870533202)
```

```r
barplot(dbinom(c(0:10),100,0.167),names.arg=c(0:10))
```
